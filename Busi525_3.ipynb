For Part 1, as T gets larger, the bias in beta hat gets smaller. The average gets closer and closer to the true beta.

For part 2, we see that as the absolute value of the correlation coefficient between u and v gets larger, then the bias in beta hat also grows.

Thus, small sample size from part 1 or correlation between u and v from part 2 can increase the bias in beta hat.

#Part 1

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

alpha = 0
beta = 0.015
sig_u = 0.053
theta = 0
sig_v = 0.044
p = 0.98
p_uv = -0.8
B = 250
```

```python
T = 120
betas = []

for b in range(B):
  x = np.zeros(T)
  r = np.zeros(T)

  for t in range (1, T):
    cov = [[sig_u**2, p_uv * sig_u * sig_v], [p_uv * sig_u * sig_v, sig_v**2]]
    u, v = np.random.multivariate_normal([0, 0], cov)

    x[t] = theta + p * x[t-1] + v
    r[t] = alpha + beta * x[t-1] + u
  x_lag = sm.add_constant(x[:-1])
  model = sm.OLS(r[1:], x_lag)
  result = model.fit()
  beta_h = result.params[1]

  betas.append(beta_h)

mean_b = np.average(betas)
_5_b = np.percentile(betas, 5)
_95_b = np.percentile(betas, 95)

print(f"Average beta: {mean_b}")
print(f"5th percentile of beta: {_5_b}")
print(f"95th percentile of beta: {_95_b}")
plt.hist(betas)
plt.axvline(x=mean_b, color='red', label=f'Mean ')
plt.axvline(x=_5_b, color='green', label=f'5th percentile ')
plt.axvline(x=_95_b, color='blue', label=f'95th percentile ')
plt.legend()
plt.show()
```

#Part 2

```python
T_values = [120, 240, 360, 480, 600, 720, 840, 960, 1080, 1200]
p_uv_values = [-0.2, -0.5, -0.8]

mean_b = {p_uv: [] for p_uv in p_uv_values}

for p_uv in p_uv_values:
    for T in T_values:
        betas = []
        for _ in range(B):
            x = np.zeros(T)
            r = np.zeros(T)
            for t in range(1, T):
                cov = [[sig_u**2, p_uv * sig_u * sig_v], [p_uv * sig_u * sig_v, sig_v**2]]
                u, v = np.random.multivariate_normal([0, 0], cov)
                x[t] = theta + p * x[t - 1] + v
                r[t] = alpha + beta * x[t - 1] + u

            x_lag = sm.add_constant(x[:-1])
            model = sm.OLS(r[1:], x_lag)
            result = model.fit()
            beta_h = result.params[1]
            betas.append(beta_h)

        mean_b[p_uv].append(np.mean(betas))

plt.figure(figsize=(10, 6))
for p_uv in p_uv_values:
    plt.plot(T_values, mean_b[p_uv], label=f'p_uv = {p_uv}')
plt.xlabel('T values')
plt.ylabel('Average Estimated beta')
plt.legend()
plt.grid(True)
plt.show()
```
